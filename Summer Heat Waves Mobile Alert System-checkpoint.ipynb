```python
import pandas as pd
import matplotlib.pyplot as plt

def identify_heat_wave(df, threshold=35, consecutive_days=3):
    df['heat_wave'] = ((df['Temp Max'] > threshold).astype(int)
                       .groupby(df['Temp Max'].gt(threshold).cumsum())
                       .transform('sum') >= consecutive_days).astype(int)
    return df

def add_temp_exceeds_column(df, threshold=28):
    df['temp_exceeds_28'] = (df['Temp Max'] > threshold).astype(int)
    return df

data = pd.read_csv('temp.csv')

data['Date'] = pd.to_datetime(data['Date'])

data.sort_values('Date', inplace=True)

data = identify_heat_wave(data)

data = add_temp_exceeds_column(data)

print(data[['Date', 'Temp Max', 'heat_wave', 'temp_exceeds_28']].head(15))

plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Temp Max'], label='Temp Max', color='b')
plt.scatter(data[data['heat_wave'] == 1]['Date'], data[data['heat_wave'] == 1]['Temp Max'], color='r', label='Heat Wave', marker='o')
plt.scatter(data[data['temp_exceeds_28'] == 1]['Date'], data[data['temp_exceeds_28'] == 1]['Temp Max'], color='g', label='Temp > 28C', marker='x')
plt.title('Temperature Over Time with Heat Waves and Temp > 28C')
plt.xlabel('Date')
plt.ylabel('Temp Max')
plt.legend()
plt.grid(True)
plt.show()
```

```python
def identify_heat_wave(df, threshold=28, consecutive_days=3):
    temp_exceeds_threshold = df['Temp Max'] > threshold
    
    group_ids = (temp_exceeds_threshold != temp_exceeds_threshold.shift()).cumsum()
    
    consecutive_counts = temp_exceeds_threshold.groupby(group_ids).transform('sum')
    
    df['heat_wave'] = (consecutive_counts >= consecutive_days).astype(int)
    
    return df

data = identify_heat_wave(data)

print(data[['Date', 'Temp Max', 'heat_wave']].head(15))
```

```python
import pandas as pd

data = pd.read_csv('temp.csv')

print("Columns in DataFrame:", data.columns)

def identify_heat_wave(df, threshold=28, consecutive_days=3):
    if 'Temp Max' not in df.columns:
        raise ValueError("Expected column 'Temp Max' not found in the DataFrame")
    
    temp_exceeds_threshold = df['Temp Max'] > threshold
    group_ids = (temp_exceeds_threshold != temp_exceeds_threshold.shift()).cumsum()
    consecutive_counts = temp_exceeds_threshold.groupby(group_ids).transform('sum')
    df['heat_wave'] = (consecutive_counts >= consecutive_days).astype(int)
    return df


if 'heat_wave' not in data.columns:
    data = identify_heat_wave(data)


print(data.head())


x = data[['Temp Max']]  # Ensure x is a 2D DataFrame
y = data['heat_wave']

# Check for missing values
print("Missing values in x:", x.isnull().sum().sum())
print("Missing values in y:", y.isnull().sum())


x = x.dropna()
y = y[x.index]  


print("Shape of x:", x.shape)
print("Shape of y:", y.shape)

# Split the data into training and testing part
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# decisiontreeclassifier model 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix

model = DecisionTreeClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
import requests
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import pickle

def fetch_weather_data(api_key, city):
    api_url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}"
    response = requests.get(api_url)
    return response.json()

def send_alert(message):
    print(message)

api_key = "b7eb98124fe9d62ebaeaf3687ed1bb8b"
city = "Mumbai"
weather_data = fetch_weather_data(api_key, city)
print(weather_data)

data = pd.read_csv('temp.csv', parse_dates=['Date'], index_col='Date')

data = data.sort_index()

data = data.asfreq('D')

model = ARIMA(data['Temp Max'], order=(5,1,0))
model_fit = model.fit()

model_fit.save('heat_wave_model.pkl')

model_fit = pickle.load(open('heat_wave_model.pkl', 'rb'))

predictions = model_fit.forecast(steps=7)

if any(predictions > 35):
    send_alert("Heat wave predicted in the next 7 days!")
```
